```python
import pandas as pd
```

```python
df = pd.read_csv("/content/Book1.csv")
```

```python
missing_values = df.isna().sum()
print(missing_values)
```

```python
print(df.describe())
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(8,6))
df.boxplot()
plt.xticks(rotation=45)
plt.title("Box Plot of Variables")
plt.show()
```

```python
print(df.head())
```

```python
region_summary = df.groupby("State_Region")["fv_tot"].agg(["mean","median","sum"])
print("region wise summary:")
print(region_summary)
```

```python
district_summary = df.groupby("District")["fv_tot"].agg(["mean","median","sum"])
print("district wise summary:")
print(district_summary)
```

```python
top_districts = district_summary.sort_values(by='sum', ascending=False).head(3)
print("\nTop Three Districts by Consumption:")
print(top_districts)
```

```python
bottom_districts = district_summary.sort_values(by='sum').head(3)
print("\nBottom Three Districts by Consumption:")
print(bottom_districts)
```
